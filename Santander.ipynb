```python
# created column for age groups

def age_group (column):
    if column['age'] <= 29 :
      return '< 30'
    if column['age'] <= 39 :
      return '30 - 39'
    if column['age'] <= 49 :
      return '40 - 49'
    if column['age'] <= 59 :
      return '50 - 59'
    if column['age'] > 60 :
      return '60 +'
```

```python
df['age_group'] = df.apply (lambda column: age_group(column), axis=1)
```

```python
def segment (column):
    if column['segmento'] == '01 - TOP' :
      return 1
    if column['segmento'] == '02 - PARTICULARES' :
      return 2
    if column['age'] == '03 - UNIVERSITARIO' :
      return 3
```

```python
df['segment'] = df.apply (lambda column: segment(column), axis=1)
```
